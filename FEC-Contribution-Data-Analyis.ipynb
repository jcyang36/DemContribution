```python
import numpy as np
import pandas as pd
import zipfile

filepath="P00000002-ALL.zip"

zf=zipfile.ZipFile(filepath)
with zf.open(filepath,mode="r") as myfile:
    headRow = next(myfile)

# read column names    
columns = [x.strip() for x in headRow.split(',')]

# process by pandas
df = pd.read_csv('P00000002-ALL.zip',names=columns[1:],compression='zip', sep=',', quotechar='"')
df.head()
```

```python
candidatels=[]
contributionSt=[]
contributionAmt=[]
datels=[]
```
